```python
# Classificacao.py
# Eh gordinho?, tem pernas curtas?, ele late?
porco1 = [1,1,0]
porco2 = [1,1,0]
porco3 = [1,1,0]
cachorro1 = [1,1,1]
cachorro2 = [0,1,1]
cachorro3 = [0,1,1]

dados = [porco1, porco2, porco3, cachorro1, cachorro2, cachorro3]
# As respostas serão chamadas de marcações, essas devem ter valores distintos
# para elementos diferentes:
# 1:porco;
# -1:cachorro.
marcacoes = [1,1,1,-1,-1,-1]

misterioso1 = [1,1,1]

# Usando um algoritimo de ML
# "Treinar um modelo": processo de aplicar um algoritmo de ML a um conjunto de dados
# Usaremos o algoritmo MultinomialNB

from sklearn.naive_bayes import MultinomialNB
modelo = MultinomialNB()

# Existem dois métodos disponíveis no modelo:
# fit(dados, marcacoes): treina um modelo
# predict([dados]): prevê os resultados para novos elementos

# Abaixo é um cachorro
modelo.fit(dados, marcacoes)
print(modelo.predict([misterioso1]))
# Abaixo é um porco
misterioso2=[1,0,0]
print(modelo.predict([misterioso2]))
# Abaixo era pra ser um cachorro
misterioso3 = [0,0,1]
print(modelo.predict([misterioso3]))
# Abaixo uma parte em que o algoritmo falha, era esperado porco
misterioso4 = [0,0,0]
print(modelo.predict([misterioso4]))

#propondo todas as entradas possiveis e vendo a taxa de acerto em % do algoritmo
# Eh gordinho?, tem pernas curtas?, ele late?

# Porco
misterioso1 = [0,0,0]
# Porco
misterioso2 = [1,0,0]
# Porco
misterioso3 = [0,1,0]
# Cachorro
misterioso4 = [0,0,1]
# Porco
misterioso5 = [1,1,0]
# Cachorro
misterioso6 = [1,0,1]
# Cachorro
misterioso7 = [0,1,1]
# Cachorro
misterioso8 = [1,1,1]

animais = [misterioso1, misterioso2, misterioso3, misterioso4, misterioso5, misterioso6, misterioso7, misterioso8]

for animal in animais:
    tipo = 0
    tipo = modelo.predict([animal])
    if tipo == 1:
        print('Porco' + " " + str(tipo))
    else:
        print('Cachorro' + " " + str(tipo))
        
print("Taxa de erro: ", str(100/6) + "%")
```
